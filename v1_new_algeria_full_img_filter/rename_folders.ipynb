```python
import os
```

```python
folders_to_rename = ["AHMEUR_BOU_AHMEUR_ab",
                     "AMER_BOUAMAR_ad", 
                     "AMER_BOUAMAR_ab", 
                     "BABARI_ad",
                     "BABARI_ab",
                     "BOUABANE_DES_AURES_ad",
                     "BOUABANE_DES_AURES_ab",
                     "ICHMOUL_ad",
                     "ICHMOUL_ab",
                     "ICHMOUL_BACHA_ab",
                     "LOULI_ab",
                     "TIZI_OUININE_ab",
                    ]

def rename_files_in_folders(folders):
    """
    Renames files in a list of folders based on the folder name and sequential number.

    Args:
        folders (list): A list of folder paths.
    """
    for folder_path in folders:
        print(f"Processing folder: {folder_path}")
        file_counter = 1
        try:
            for filename in os.listdir(folder_path):
                old_file_path = os.path.join(folder_path, filename)
                if os.path.isfile(old_file_path):
                    folder_name = os.path.basename(folder_path)
                    _, file_extension = os.path.splitext(filename)
                    new_filename = f"{folder_name}_{file_counter}{file_extension}"
                    new_file_path = os.path.join(folder_path, new_filename)
                    try:
                        os.rename(old_file_path, new_file_path)
                        print(f"  Renamed '{filename}' to '{new_filename}'")
                        file_counter += 1
                    except Exception as e:
                        print(f"  Error renaming '{filename}': {e}")
        except FileNotFoundError:
            print(f"Error: Folder not found: {folder_path}")
        except Exception as e:
            print(f"Error processing folder '{folder_path}': {e}")

# Example usage:
folders_to_rename = folders_to_rename
rename_files_in_folders(folders_to_rename)
        


```
