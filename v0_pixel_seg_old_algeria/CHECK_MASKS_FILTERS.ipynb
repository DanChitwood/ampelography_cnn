```python
import cv2
import matplotlib.pyplot as plt

img = "335_GENOTYPE 2_ID36"

# Load the image in grayscale mode
image = cv2.imread("./masks/"+img+".png", cv2.IMREAD_GRAYSCALE)

# Display the image using matplotlib
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

```python
import numpy as np
import cv2

# Assuming filter1_img and filter2_img are your filtered grayscale images
sato_img = cv2.imread('./sato/'+img+".png", cv2.IMREAD_GRAYSCALE)
meijering_img = cv2.imread('./meijering/'+img+".png", cv2.IMREAD_GRAYSCALE)
frangi_img = cv2.imread('./frangi/'+img+".png", cv2.IMREAD_GRAYSCALE)
hessian_img = cv2.imread('./hessian/'+img+".png", cv2.IMREAD_GRAYSCALE)
```

```python
# Addition
add_img = sato_img.astype(np.int32)+meijering_img.astype(np.int32)+frangi_img.astype(np.int32)+hessian_img.astype(np.int32)

# Normalize to 0-255
add_img = np.clip(add_img, 0, 255).astype(np.uint8)

plt.imshow(add_img)
```

```python
# Multiply
add_img = sato_img.astype(np.int32)*meijering_img.astype(np.int32)*frangi_img.astype(np.int32)*hessian_img.astype(np.int32)

# Normalize to 0-255
add_img = np.clip(add_img, 0, 255).astype(np.uint8)

plt.imshow(add_img)
```
